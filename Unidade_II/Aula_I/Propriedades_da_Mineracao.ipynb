**Tabelas de decisão**
> É uma maneira de expressar, em forma de tabela, qual o conjunto de condições que é necessário ocorrer para que um determinado conjunto de ações deva ser executado.
> 
> O ponto principal de uma tabela de decisão é a regra de decisão que define o conjunto de ações a ser tomadas, a partir de um conjunto de condicoes.
> 
> Regras de decisão:
>> Caso o cartão seja considerado inválido;
>>
>> Se o cartão for válido mas a senha não for válida.
>>
>> Se ambos forem válidos mas o saldo é insuficiente para saque.
>>
>> Quando o cartão for válido, a senha for válida, houver saldo suficiente, o saque é realizado.

**O que é WEKA?**
> WEKA é um conjunto de mineração de dados desenvolvida na Universidade de Waikato.
> 
> WEKA significa Ambiente para Análise de Conhecimento.
> 
> WEKA inclui tudo o que é necessário para gerar e aplicar modelos de mineração de dados.
>> Abrange todas as principais tarefas de mineração de dados.
>>
>> Inclui ferramentas para pré-processar e visualizar dados.
>>
>> Inclui várias interfaces.

**Por que usar WEKA?**
> Vantagens:
>> Fácil de aprender.
>>
>> Gratuito e de código aberto.
>>
>> Fácil de baixar e roda em muitas plataformas.
>>
>> Não requer conhecimento de programação.

> Desvantagens:
>> Não é tão flexível quanto outras ferramentas baseadas em programação.
>>
>> Ferramentas baseadas em Python e R tem um grande ecossistema.

**Formato de arquivo ARFF**
> ARFF = Formato de arquivo de relação de atributos.
>
> Os arquivos ARFF incluem duas partes principais:
>> Especificação de funcionalidades.
>> Dados reais.
>
> Arquivos são simples separados por virgulas.
> 
> Pode ser dividido em duas partes:
>> 1. Especificação de funcionalidades.
>> 2. Dados reais.

# Videoaula: Geração de informação para ações práticas

**Algoritmo a Priori**
> É o método mais conhecido para a mineração de regras de associação e emprega busca em profundidade, ele gera conjuntos de itens candidatos de k elementos a partir de conjuntos de itens com k-1 elementos.
> 
> Os itens candidatos não frequentes são eliminados, e toda a base de dados é rastreada e os conjuntos de itens frequentes, obtidos a partir de conjuntos de itens candidatos.
> 
> Algoritmo de Suporte a Associação.
> 
> Uma forma de reduzir o custo computacional dos algoritmos de mineração de regras de associação é desacoplar os requisitos de suporte e confiança mínima das regras.
> 
> Uma estratégia comum adotada pelos algoritmos de mineração de regras de associação consiste em decompor o problema em duas subtarefas.
> 
> Geração do conjunto de itens frequentes.
> 
> Geração das regras.
>
> Contagem de suporte.
>
> Geração dos conjuntos-K
>
> Determinação dos conjuntos frequentes.
>
> Critério de parada.
>
> O algoritmo percorre os conjuntos de itens frequentes iterativamente em ordem crescente de tamanho, gerando e testando cada conjunto até encontrar os frequentes.
